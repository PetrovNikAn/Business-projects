```python
# определим топ-10 самых нерентабельных партнеров по показателю NPV по имеющимся данным:
unprofitable_partner = data_SBER.groupby('ИД партнера')['NPV, руб'].sum().sort_values(ascending = True)
unprofitable_partner.head (10)
```

```python
# посмотим тип деятельности партнеров, вошедших в перечень наименее рентабельных:
top_unprof_partner = ['592***','485***', '409***', '317***', '570***', '457***', '463***', '576***', '122***', '138***']

# для безошибочной работы кода заменим русскоязычное название графы с "ИД парнетров" и "МСС" на англоязычное:
data_SBER.rename(columns = {'ИД партнера':'id_partners', 'МСС':'mcc'}, inplace = True)
data_temp = data_SBER.query('id_partners in @top_unprof_partner').pivot_table(index = 'mcc', columns = 'ГОСБ', values = 'NPV, руб', aggfunc='sum')
data_temp
```

```python
# проверим, в каком ГОСБ наблюдается наименьший показатель NPV по топ-10 парнерам:
data_temp_gosb = data_SBER.query('id_partners in @top_unprof_partner').groupby('ГОСБ')['NPV, руб'].sum()
print(data_temp_gosb)
data_temp_gosb.plot(kind='bar')
```

В Саратовском ГОСБ№8622 замечен наихудший показатель NPV

```python
# проверим, какой вид деятельности наименее рентабелен по топ-10 партнерам:
data_temp_okved = data_SBER.query('id_partners in @top_unprof_partner').groupby('mcc')['NPV, руб'].sum().sort_values()
print(data_temp_okved)
data_temp_okved.head(10).plot(kind='bar')
```
